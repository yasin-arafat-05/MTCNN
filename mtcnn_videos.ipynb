# This section of code for detecting faces from a video:

<br>
<br>

```python
import cv2 as cv
from mtcnn import MTCNN

detector = MTCNN()
```


<br>
<br>

# 1) From my laptop camera: 

<br>
<br>


```python
cap = cv.VideoCapture(0)

while cap.isOpened():
    ret,frame = cap.read()
    
    # ret is telling frame is comming or not(boolean value): 
    if not ret:
        break
    
    
    #detect face using mtcnn:
    output = detector.detect_faces(frame)
    
    for single_output in output:
        x,y,width,height = single_output['box']
        cv.rectangle(img=frame,pt1=(x,y),pt2=(x+width,y+height),color=(255,0,0),thickness=5)
    
    # opencv read in bgr:
    frame = cv.cvtColor(frame,cv.COLOR_BGR2RGB)
    
    # Flip the frame horizontally (axis=1)
    frame = cv.flip(frame, 1)
    
    cv.imshow('output',frame)
    if cv.waitKey(1) == ord('q'):
        break
    
cap.release()
cv.destroyAllWindows()
    
```

```python
cap = cv.VideoCapture('images/videoplayback.webm')

while cap.isOpened():
    ret,frame = cap.read()
    
    # ret is telling frame is comming or not(boolean value): 
    if not ret:
        break
    
    
    #detect face using mtcnn:
    output = detector.detect_faces(frame)
    
    for single_output in output:
        x,y,width,height = single_output['box']
        cv.rectangle(img=frame,pt1=(x,y),pt2=(x+width,y+height),color=(255,0,0),thickness=5)
    
    # opencv read in bgr:
    frame = cv.cvtColor(frame,cv.COLOR_BGR2RGB)
    
    # Flip the frame horizontally (axis=1)
    frame = cv.flip(frame, 1)
    
    frame = cv.resize(frame,(720,480))
    cv.imshow('output',frame)
    if cv.waitKey(1) == ord('q'):
        break
    
cap.release()
cv.destroyAllWindows()
    
```
