# Facial Land Mark: 

<br>
<br>

---
---

<br>
<br>


```python

left_eyeX,left_eyeY = output[0]['keypoints']['left_eye']
right_eyeX,right_eyeY = output[0]['keypoints']['right_eye']
mouth_leftX,mouth_leftY = output[0]['keypoints']['mouth_left']
mouth_rightX,mouth_rightY = output[0]['keypoints']['mouth_right']
noseX,noseY = output[0]['keypoints']['nose']
```

```python
cv.circle(img,center=(left_eyeX,left_eyeY),radius=5,color=(0,255,0),thickness=-1)
cv.circle(img,center=(right_eyeX,right_eyeY),radius=5,color=(0,255,0),thickness=-1)
cv.circle(img,center=(mouth_leftX,mouth_leftY),radius=5,color=(0,255,0),thickness=-1)
cv.circle(img,center=(mouth_rightX,mouth_leftY),radius=5,color=(0,255,0),thickness=-1)
cv.circle(img,center=(noseX,noseY),radius=5,color=(0,255,0),thickness=-1)
plt.imshow(img)
```

# Multiple Face In A Single Photo: 

<br>
<br>

---
---

<br>
<br>


```python
multi = cv.imread('images/bond.jpg')
multi = cv.cvtColor(multi,cv.COLOR_BGR2RGB)
plt.imshow(multi)
```

```python
multi_face = decetor.detect_faces(img=multi)
multi_face
```

```python

for i in multi_face:
    
    left_eyeX,left_eyeY = i['keypoints']['left_eye']
    right_eyeX,right_eyeY = i['keypoints']['right_eye']
    mouth_leftX,mouth_leftY = i['keypoints']['mouth_left']
    mouth_rightX,mouth_rightY = i['keypoints']['mouth_right']
    noseX,noseY = i['keypoints']['nose']

    cv.circle(multi,center=(left_eyeX,left_eyeY),radius=5,color=(0,255,0),thickness=-1)
    cv.circle(multi,center=(right_eyeX,right_eyeY),radius=5,color=(0,255,0),thickness=-1)
    cv.circle(multi,center=(mouth_leftX,mouth_leftY),radius=5,color=(0,255,0),thickness=-1)
    cv.circle(multi,center=(mouth_rightX,mouth_leftY),radius=5,color=(0,255,0),thickness=-1)
    cv.circle(multi,center=(noseX,noseY),radius=5,color=(0,255,0),thickness=-1)
    
    x,y,height,width = i['box']
    cv.rectangle(img=multi,pt1=(x,y),pt2=(x+width,y+height),color=(255,0,0),thickness=5)
    
plt.imshow(multi)
```
